We see the data contains of the number, the name, sometimes an extra name, two types and some stats. One way to go about getting the data sorted from each string would be to apply regex. Getting the number and the stats would be easy, but when the pokemon sometimes has an extra name it gets more difficult. Therefor I've broken the problem down to smaller bits, getting one type of information at a type, well knowing that it can be made smarter with a single regex expression.

### Pokemon Name

Therefor I've decided to make a function that does that, getting the name of the pokemons.

```python
def getPokemonName(pokemon):
    #The pokemon normal name is found using the class ent-name
    pokemonName = pokemon.find("a", class_="ent-name").text
    
    #The pokemon special name is found using the class aside
    pokemonSpecialName = pokemon.find("small", class_="aside")
    
    #Only if the special name exists will it be casted to a string
    if pokemonSpecialName != None:
        pokemonSpecialName = pokemonSpecialName.text 
    return (pokemonName,pokemonSpecialName)
```

```python
totalPokemonNames = []
for pokemon in pokedex.findAll('tr')[1:]:
    totalPokemonNames.append(getPokemonName(pokemon))
totalPokemonNames[:6]
```

It works!

### Pokemon Types

The next data from the string contains the type of Pokemon, which can be either single or two types.

```python
def getPokemonTypes(pokemon):
    #For every pokemon's type using the class "cell-icon"
    for pokemonType in pokemon.findAll('td', class_="cell-icon"):
        #Regex to split the string by upperCase
        pokemonTypes = re.findall('[A-Z][^A-Z]*', pokemonType.text)
        #If the lenght is 1 type two is set two None
        if len(pokemonTypes) == 1:
            pokemonType1 = pokemonTypes[0]
            pokemonType2 = None
        # Else the two types are returned
        else:
            pokemonType1 = pokemonTypes[0]
            pokemonType2 = pokemonTypes[1]
    return (pokemonType1,pokemonType2)
```

```python
totalPokemonTypes = []
for pokemon in pokedex.findAll('tr')[1:]:
    totalPokemonTypes.append(getPokemonTypes(pokemon))
totalPokemonTypes[:6]
```

It works!

### Pokemon Stats

Now we only need to get the pokemon stats. Which can be acquired easily by finding the "num" class and the "num-total" class

```python
def getPokemonStats(pokemon):
    #The total stats is easy to get by getting the "num-total" class, another way could be to add the other stats
    pokemonTotalStats = int(pokemon.find('td', class_="num-total").text)
    #For everyother stats the class is called the same ("nam"), but they come in order
    #Using counter and enumurate gets the data for each position
    for counter, pokemonStats in enumerate(pokemon.findAll('td', class_="num")): 
        if counter == 1:
            pokemonHP = int(pokemonStats.text)
        elif counter == 2:
            pokemonAttack = int(pokemonStats.text)
        elif counter == 3:
            pokemonDefense = int(pokemonStats.text)
        elif counter == 4:
            pokemonSPAttack = int(pokemonStats.text)
        elif counter == 5:
            pokemonSPDefense = int(pokemonStats.text)
        elif counter == 6:
            pokemonSpeed = int(pokemonStats.text)
    #Packing it up in a list before returing it
    pokemonStats = [pokemonTotalStats,pokemonHP,pokemonAttack,pokemonDefense,pokemonSPAttack,pokemonSPDefense,pokemonSpeed]
    return pokemonStats
```

```python
totalPokemonStats = []
for pokemon in pokedex.findAll('tr')[1:]:
    totalPokemonStats.append(getPokemonStats(pokemon))
totalPokemonStats[:6]
```

It works!

###Combinging the data

We want a dataframe which contain: 'Name','SpecialName','TypeOne','TypeTwo','Total','HP','Attack','Defense','Sp.Atk','Sp.Def' and Speed' Therefor we build one empty first

```python
columns = ['Name','SpecialName','TypeOne', 'TypeTwo','Total','HP','Attack','Defense','Sp.Atk','Sp.Def','Speed']
df = pd.DataFrame(columns=columns)
df
```

Now we simply call each and every method we just made, combine it into a single row of dataframe and return it.

```python
def getPokemonInfo(pokemon):
    
    pokemonName,pokemonSpecialName = getPokemonName(pokemon)
    pokemonType1,pokemonType2 = getPokemonTypes(pokemon)
    
    pokemonStats = getPokemonStats(pokemon)
    #combining into one big list
    pokemonCombined = [pokemonName,pokemonSpecialName,pokemonType1,pokemonType2] + pokemonStats
    #Creating a dataframe row from the list and the columns
    pokemonRow = pd.DataFrame([pokemonCombined], columns=columns)
    return(pokemonRow)
```

To test it all, we use our pokedex data and the lopp for every pokemon in it. We rename index as number as the number of the pokemon equals the index +1

```python
for pokemon in pokedex.findAll('tr')[1:]:
    df = df.append(getPokemonInfo(pokemon),ignore_index=True)
df.index.rename('Number', inplace=True)
df.head()
```

### Cleaning the data

Now the data is in a dataframe it's time to see what type it is and parse it to numerical values if needed

```python
df.info(memory_usage='deep')
```

Every stat value should be an int, every type value should be an category and the name should be objects as they are 99% different in all


```python
statList = ['Total',
 'HP',
 'Attack',
 'Defense',
 'Sp.Atk',
 'Sp.Def',
 'Speed']
df[statList] = df[statList].astype(np.uint16)
df['TypeOne'] = df['TypeOne'].astype(dtype="category")
df['TypeTwo'] = df['TypeTwo'].astype(dtype="category")
```

We have greatly reduced the size of the data and made most of it numerical and easy to compute

### Statistical Analysis

Now that we have a clean Data Frame, let's do some analysis. We wish to find the best (and worst) Pokemon for each stat.

```python
def max_stats(df, col_list):
    '''Get Pokemon highest value of the column in the Data Frame'''
    message = ''
    for col in col_list:
        highStat = df[col].max()
        highName = df[df[col]==df[col].max()]['Name'].values[0]
        lowStat = df[col].min()
        lowName = df[df[col]==df[col].min()]['Name'].values[0]
        
        introMessage = 'The ' + col + ' stat!'
        print(introMessage)
        print(df[col].describe())

        highestMessage = highName+' has the greatest '+col+' of '+str(highStat)+'.'
        lowestMessage = lowName+' has the worst '+col+' of '+str(lowStat)+'.'
        print(highestMessage)
        print(lowestMessage)
        df[col].plot(kind='hist',title =col+' Stat')
        plt.show()
```

```python
stats=['Attack', 'Defense','HP', 'Sp.Atk','Sp.Def','Speed','Total']
max_stats(df, stats)
```

As usual, Mewtwo takes the lead for multiple categories!

Now that we have seen the information on abilities, let's dig deeper!

For out next analysis, we will the effect of Pokemon type on abilities.

```python
#Creating new Data Frame where the Type values are separated from the list
newDict={}
stats_col=["#","Name","Total","HP","Attack","Defense","Sp. Atk","Sp. Def","Speed"]

#Collecting the list of Type for each Pokemon
Dict['Type']=df['Type'].values

#Creating an empty list for each key (column)
for col in stats_col:
    newDict[col]=[]
    newDict['Type']=[]

#Populating each the dictionary value (empty list) with data 
for row in range(len(Dict['#'])):
    for t in Dict['Type'][row]:
        for col in stats_col:
            #Append all columns except Type to the new dictionary
            newDict[col].append(Dict[col][row])
        #Appending the Type separately for each Pokemon  in the new dictionary
        newDict['Type'].append(t)
        
#Convert dictionary to a Data Frame
new_df = pd.DataFrame(newDict)
```

```python
new_df.head()
```

Now that we have a new Data Frame restructured based on Type, let's see the unique Pokemon types.

```python
types=new_df['Type'].unique()
print types
```

```python
#Colours to cycle through when plotting the hbar graph
my_colors = list(six.iteritems(colors.cnames))
my_colors = list(islice(cycle(my_colors), None, len(new_df)))

def barh_stats():
    '''Plot hbar of mean and std. of each attribute of Pokemon Type'''
    i=0
    plt.figure(figsize=(15,5))
    plt.suptitle('Statistics', fontsize=15)
    
    
    #Cycle through each pokemon Type
    for t in types:
        
        #Iterate i value to change colour to my_colors[i]
        i+=1   
        
        #Plotting Mean
        plt.subplot(121)
        plt.title('Mean')
        new_df[new_df['Type']==t].mean().plot(kind='barh', color=my_colors[i])
        
        #Plotting Standard Deviation
        plt.subplot(122)
        plt.title('Standard Deviation')
        new_df[new_df['Type']==t].std().plot(kind='barh', color=my_colors[i])
    
    #Add list of Pokemon Type to legend
    plt.legend(types,bbox_to_anchor=(1.3, 1.1))
```

```python
barh_stats()
```

Based on the graph, we can see that Ice types are better on average based on Total and Defense. 

Dark Types tend to have the grated variation in strength based on Total  and Sp. Attack while the greatest variation for Defense is among Ice types.

###GUI (aka Pokedex)

Now for the fun bit. We will try building a PokeDex. 

For this we will use the <b>PokemonData.json</b> file we have created before and look at the data using a Python GUI.

```python
import pandas as pd
import sys, urllib2
import requests
from PyQt4 import QtGui, QtCore

class PokeDex(QtGui.QWidget):
    
    def __init__(self):
        super(PokeDex, self).__init__()
        
        self.initUI()
        
    def initUI(self):
        '''Initial UI'''
        
        #Grid Layout
        self.grid = QtGui.QGridLayout()
        self.setLayout(self.grid)
        
        #Parse JSON for DataFrame
        self.df = pd.read_json('PokemonData.json')
        self.df = self.df.set_index(['#'])
        
        #Drop Down
        self.dropdown = QtGui.QComboBox(self)
        self.names = self.df['Name'].values
        self.dropdown.addItems(self.names)
        self.grid.addWidget(self.dropdown, 0,0,1,1)
        
        #Search Button
        self.btn = QtGui.QPushButton('Search', self)
        self.btn.clicked.connect(self.runSearch)      
        self.grid.addWidget(self.btn, 0,1,1,1)
        
        #Image
        self.img = QtGui.QLabel(self)
        self.grid.addWidget(self.img, 1,1,1,1)
        
        #Data
        self.label = QtGui.QLabel(self)
        self.label.setAlignment(QtCore.Qt.AlignCenter)
        self.label.setText('\nName:\n\nType:\n\nHP:\n\nAttack\n\nSp. Attack\n\n Defense:\n\nSp. Defense:\n\nSpeed:\n\nTotal:')
        self.label.setAlignment(QtCore.Qt.AlignLeft)
        self.grid.addWidget(self.label, 1,0,1,1)
        
        #Customize Widgets
        self.resize(500, 250)
        self.center()
        self.setWindowTitle('PokeDex')    
        self.show()
        
    def runSearch(self):
        '''Event for run button'''
        
        #Parse value
        index = self.dropdown.currentIndex()
        val= self.names[index]
        cond = self.df['Name']== val
        
        #Image
        base = 'http://img.pokemondb.net/artwork/'
        img_url = base + val.lower() + '.jpg'
        data = urllib2.urlopen(img_url).read()
        image = QtGui.QImage()
        image.loadFromData(data)
        self.img.setPixmap(QtGui.QPixmap(image))
        
        #Set values
        name = 'Name:\t\t\t'+val+'\n\n'
        ty = 'Type:\t\t\t'+ ', '.join(self.df[cond]['Type'].values[0])+'\n\n'
        hp = 'HP:\t\t\t'+ str(self.df[cond]['HP'].values[0])+'\n\n'
        atk = 'Attack:\t\t\t'+str(self.df[cond]['Attack'].values[0])+'\n\n'
        satk = 'Sp. Attack:\t\t'+str(self.df[cond]['Sp. Atk'].values[0])+'\n\n'
        deff = 'Defense:\t\t\t'+str(self.df[cond]['Defense'].values[0])+'\n\n'
        sdef = 'Sp. Defense:\t\t'+str(self.df[cond]['Sp. Def'].values[0])+'\n\n'
        speed = 'Speed:\t\t\t'+str(self.df[cond]['Speed'].values[0])+'\n\n'
        total = 'Total:\t\t\t'+str(self.df[cond]['Total'].values[0])+'\n\n'
        
        #Add text
        final = name+ty+hp+atk+satk+deff+sdef+speed+total
        self.label.setText(final)
        
    def center(self):
        '''Center Widget on screen'''
        qr = self.frameGeometry()
        cp = QtGui.QDesktopWidget().availableGeometry().center()
        qr.moveCenter(cp)
        self.move(qr.topLeft())
        
        
def main():
    '''Codes for running GUI'''
    
    #Create Application object to run GUI
    app = QtGui.QApplication(sys.argv)
    app.aboutToQuit.connect(app.deleteLater)
    
    #Run GUI
    gui = PokeDex()
    #Exit cleanly when closing GUI
    sys.exit(app.exec_())


if __name__ == '__main__':
    main()
```
